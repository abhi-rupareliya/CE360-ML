## Line Chart


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
# Python Program to illustrate Linear Plotting
# year contains the x-axis values and e-india & e-bangladesh are the y-axis values for plotting

year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]

# Setting figure size
fig = plt.figure(figsize=(8, 6))
# plotting of x-axis(year) and y-axis(power consumption) with different colored labels of two countries

plt.plot(year, e_india, color='orange', label='India', marker='*',
         markersize=12)  # marker ='o', markersize = 12,

plt.plot(year, e_bangladesh, color='g', label='Bangladesh',
         linestyle='dashed', linewidth=2)  # linestyle ='dashed', linewidth = 2,

font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'serif', 'color': 'darkred', 'size': 15}
plt.grid()
# naming of x-axis and y-axis
plt.xlabel('Years', fontdict=font1)
plt.ylabel('Power consumption in kWh', fontdict=font1)

# naming the title of the plot
plt.title(
    'Electricity consumption per capita of India and Bangladesh', fontdict=font2)

pop_a = mpatches.Patch(color='#5DC83F', label='India')
pop_b = mpatches.Patch(color='#CE5D45', label='Bangladesh')
plt.legend(handles=[pop_a, pop_b], frameon=True, labelspacing=2, ncol=2)


# plt.legend()
# The attribute Loc in legend() is used to specify the location of the legend.Default value of loc is loc=best (upper left).
# The strings upper left, upper right, lower left, lower right place the legend at the corresponding corner of the axes/figure.

# The attribute bbox_to_anchor=(x, y) of legend() function is used to specify the coordinates of the legend,
# and the attribute ncol represents the number of columns that the legend has.Its default value is 1.

# plt.legend(["Coutry-1", "Country-2"], loc ="lower right")
# plt.legend(bbox_to_anchor =(0.75, 1.25), ncol = 2)
# plt.fill_between(e_india, e_bangladesh, year, color='green', alpha=0.5)
plt.show()
```

## Subplot


```python
# plot1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 2 row, 2 columns, and this plot is the first plot.
plt.subplot(2, 2, 1)
plt.title("PLOT 1")
plt.plot(x, y)

# plot2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 2, 2)
plt.title("PLOT 2")
plt.plot(x, y)

# plot 3
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 2, 3)
plt.title("PLOT 3")
plt.plot(x, y)

# plot 4
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 2, 4)
plt.title("PLOT 4")
plt.plot(x, y)

plt.suptitle("SUBPLOTS")


# spacing betn subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.8,  # spacing between horizontal elements
                    hspace=0.5)  # spacing between vertical elements


# plt.tight_layout()  # way 2
```

```python
# 2 subplots in 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

x1 = ['Telugu', 'Hindi', 'English', 'Maths', 'Science', 'Social']
y1 = [45, 34, 30, 45, 50, 38]
y2 = [36, 28, 30, 45, 38, 48]

# Labels to use in the legend for each line
labels = ["in 2019", "in 2020"]

# Title for subplots
fig.suptitle(
    'Number of Students passed in each subject from a class in 2019 & 2020', fontsize=14)

# Creating the sub-plots.
l1 = ax1.plot(x1, y1, color="green")
l2 = ax2.plot(x1, y2, color="blue")

ax1.set_yticks(np.arange(0, 51, 5))
ax2.set_yticks(np.arange(0, 51, 5))

ax1.set_ylabel('Number of students', fontsize=25)


fig.legend([l1, l2], labels=labels, loc="upper right")

# Adjusting the sub-plots
# plt.subplots_adjust(right=0.8)

plt.show()
```

## Bar Chart

## matplotlib.pyplot.bar(x, width=0.8 , height, bottom=None, align='center', data=None, \*\*kwargs)

# Parameters of Multiple-bar Chart

# x: indicate the bars' x-coordinates.

# height: The height of the bars is determined by the y-coordinates.

# width: the width of the bars must be specified.

# bottom: the y coordinates of the bar bases should be specified.

# align: The bars must be aligned.


```python
# creating the dataset
data = {'C': 20, 'C++': 15, 'Java': 30, 'Python': 35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10, 5))

# creating the bar plot
plt.bar(courses, values, color=['maroon', "green", "blue", "red"], width=0.4)
# plt.barh(courses, values,color ='maroon',height=0.6) # horizontal bars

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

## Stacked Bar Chart


```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

plt.bar(x, y1, color='red')
plt.bar(x, y2, bottom=y1, color='green')
plt.bar(x, y3, bottom=y1+y2, color='blue')
plt.bar(x, y4, bottom=y1+y2+y3, color='orange')
plt.title("Stacked Bar chart")
```
